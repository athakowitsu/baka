```python
import matplotlib.pyplot as plt
product1 = [20,40,60,80,100]
product2 = [15,30,10, 20,60] 
month=[1,2,3,4,5] 
```

```python
#plt.plot(month,product2, color = 'm',
#marker="o",linestyle= '--')
plt.plot(month,product2,"mo--")
plt.plot(month,product1,"ro--")

plt.xlabel('month',size=20,color='red',backgroundcolor='y')
plt.ylabel('sales',size=20,color='blue') 

config={'size':20,'color':'g'}
plt.title('Top Sale 2020',config,loc='left')

plt.legend(["baka","baki"],loc=3)
```
